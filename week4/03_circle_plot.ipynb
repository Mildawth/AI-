```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-1, 1, 200) # 400 dots in total
y1 = np.sqrt(1-x**2)
y2 = -np.sqrt(1-x**2)

plt.figure(figsize=(4, 4))
plt.plot(x, y1, color='b')
plt.plot(x, y2, color='b')
```

```python
import matplotlib.pyplot as plt
import numpy as np

theta = np.linspace(0, 2*np.pi, 200) # 200 dots in total
x = np.cos(theta)
y = np.sin(theta)

plt.figure(figsize=(4, 4))
plt.plot(x, y, color='b')
```
